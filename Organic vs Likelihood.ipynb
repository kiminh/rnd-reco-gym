# _RecoGym_: Organic vs Likelihood

# Please note this is a slow notebook to run

The _`Agent`_ that uses _Organic_ _`Event`_ counter shows quite good results in predicting a click.
In this notebook you shall find how that _`Agent`_ performs when the amount of _`Products`_ flips in the _`Environment`_ is big enough to make such kind of predictions hard.

**Note:** the evaluation of the data provided in that notebook takes some time.

Below you shall find the code that initialises the _`Environment`_.

```python
import gym
import matplotlib.pyplot as plt

%matplotlib notebook
%config InlineBackend.figure_format = 'retina'
plt.rcParams['figure.figsize'] = [6, 3]

from recogym import build_agent_init, env_1_args, gather_agent_stats, plot_agent_stats

RandomSeed = 42

TrainingDataSamples = (100, 500, 1000, 2000, 3000, 5000, 8000, 10000, 13000, 14000, 15000)
TestingDataSamples = 15000
StatEpochs = 10
StatEpochsNewRandomSeed = True

std_env_args = {
    **env_1_args,
    'random_seed': RandomSeed,
}

env = gym.make('reco-gym-v1')
```

## _With_ Product Flips

In this case, there is a maximum of _`Products`_ flips in the _`Environment`_ setup.

The code that imports _`Agents`_.

```python
from recogym.agents import OrganicUserEventCounterAgent, organic_user_count_args
from recogym.agents import LogregPolyAgent, logreg_poly_args
```

```python
agent_inits = {
    **build_agent_init('User Organic', OrganicUserEventCounterAgent, organic_user_count_args),
    **build_agent_init('LogReg Poly', LogregPolyAgent, logreg_poly_args),
}
```

Gathering statistics...

```python
agent_stats01 = gather_agent_stats(
    env,
    std_env_args,
    {
        'num_products': 10,
        'number_of_flips': 5,
    },
    agent_inits,
    TrainingDataSamples,
    TestingDataSamples,
    StatEpochs,
    StatEpochsNewRandomSeed
)
```

Plotting results...

```python
plot_agent_stats(agent_stats01)
```

# _Without_ Product Flips

This time, without any _`Product`_ flips.

```python
agent_stats02 = gather_agent_stats(
    env,
    std_env_args,
    {
        'num_products': 10,
        'number_of_flips': 0,
    },
    agent_inits,
    TrainingDataSamples,
    TestingDataSamples,
    StatEpochs,
    StatEpochsNewRandomSeed
)
```

```python
plot_agent_stats(agent_stats02)
```

# Resolution

_Logistic Regression_ with a polynomial feature set
* _**does**_ outperform _User Organic Counter_ _**when**_ there are some _`Product`_ flips in the _`Environment`_, i.e. the probability of the click has _**no**_ direct relation to the amount of viewed _`Products`_;
* and, it has almost the same _Click Through Rate_ _**when**_ there is no _`Product`_ flips.
